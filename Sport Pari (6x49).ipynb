```python
#import pandas and numpy
import pandas as pd 
import numpy as np
```

### Link to all games results
https://6x49.sportloto.by/media/results.xlsx

```python
data = pd.read_excel('649.xlsx') #read data to dataframe
```

```python
data.head() #get header of results table
```

```python
data1 = data.loc[:,'Дата':'Шар 6'] #create a new dataframe with 7 first columns
```

```python
data1.head()#get 5 first rows with headers
```

```python
data1.columns = ['Date', 'Ball 1', 'Ball 2', 'Ball 3', 'Ball 4', 'Ball 5', 'Ball 6']#rename columns
```

```python
data1.head()#get 5 first rows with headers
```

```python
new_data =pd.DataFrame()#create a new dataframe
```

```python
new_data ['Ball 1'] = data1['Ball 1'].value_counts(sort=False)
new_data ['Ball 2'] = data1['Ball 2'].value_counts(sort=False)
new_data ['Ball 3'] = data1['Ball 3'].value_counts(sort=False)
new_data ['Ball 4'] = data1['Ball 4'].value_counts(sort=False)
new_data ['Ball 5'] = data1['Ball 5'].value_counts(sort=False)
new_data ['Ball 6'] = data1['Ball 6'].value_counts(sort=False)
```

```python
new_data['Total'] = new_data.sum(axis = 1)#sum all counts
```

```python
new_data.head()#get 5 first rows with headers
```

```python
new_data.sort_values(by ='Total', ascending = False)#sort numbers by frequency
```
